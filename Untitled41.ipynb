```python
mydf_pivot = mydf.pivot(index='productid', columns='UserId', values='frequency').fillna(0)
mydf_matrix = csr_matrix(mydf_pivot.values)
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric='cosine',algorithm='brute',n_neighbors=11)
model_knn.fit(mydf_matrix)
```

```python
query_index = np.random.choice(mydf_pivot.shape[0])
distance, indices = model_knn.kneighbors(mydf_pivot.iloc[query_index,:].reshape(1,-1),n_neighbors=11)
recom_arr = []
for i in range(0, len(distance.flatten())):
    if i==0:
        print('Recommendations for {0}:\n'.format(mydf_pivot.index[query_index]))
    else: 
        #print('{0}: {1}, with distance of {2}:'.format(i, mydf_pivot.index[indices.flatten()[i]], distance.flatten([i])))
        recom_arr.append(mydf_pivot.index[indices.flatten()[i]])
        
print(recom_arr)
```

```python
test_product_rec = []
test_product_ids = []
ran = mydf_pivot.shape[0]
for i in range(0,ran):
    distance, indices = model_knn.kneighbors(mydf_pivot.iloc[i,:].reshape(1,-1),n_neighbors=11)
    recom_arr = []
    for j in range(0, len(distance.flatten())):
        if j==0:
            test_product_ids.append(mydf_pivot.index[i])
        else: 
            recom_arr.append(mydf_pivot.index[indices.flatten()[j]])
    test_product_rec.append(recom_arr)
```
